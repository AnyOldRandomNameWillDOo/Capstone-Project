```python
import streamlit as st
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
import plotly.io as pio
```

```python
df = pd.read_csv('CSVFiles/FullData.csv', index_col='SEQN')
```

```python
df.head()
```

```python
df.info()
```

```python
for col in df.columns:
    print(col)
    display(df[col].unique())
```

```python
df.gender.replace({0: 'Male', 1:'Female'}, inplace=True)
```

```python
df.household_income.replace({0:'Missing', 
                             1:'Below $5K', 
                             2:'Below $10K', 
                             3:'Below $15K', 
                             4:'Below $20K', 
                             5:'Below $25K', 
                             6:'Below $35K', 
                             7:'Below $45K', 
                             8:'Below $55K', 
                             9:'Below $65K', 
                             10:'Below $75K',
                             11:'$75K+',
                             12:'Over $20K', 
                             13:'Under $20K', 
                             14:'Below $100K', 
                             15:'$100K+'}, inplace=True)
```

```python
df.current_smoker.replace({0:'Missing', 
                           1:'Every Day', 
                           2:'Some Days', 
                           3:'Never'}, inplace=True)
```

```python
df.cocaine_number_uses.replace({0:'Missing', 
                                1:'Once', 
                                2:'2-5', 
                                3:'6-19', 
                                4:'20-49', 
                                5:'50-99', 
                                6:'110+'}, inplace=True)
```

```python
df.meth_number_uses.replace({0:'Missing', 
                                1:'Once', 
                                2:'2-5', 
                                3:'6-19', 
                                4:'20-49', 
                                5:'50-99', 
                                6:'110+'}, inplace=True)
```

```python
yes_no_cols = list(df.columns)
temp_list = [0, 1]
for col in df.columns:
    x = list(sorted(df[col].unique()))
    if x != temp_list:
        yes_no_cols.remove(col)
```

```python
yes_no_cols
```

```python
for col in yes_no_cols:
    df[col].replace({0:'No', 1:'Yes'}, inplace=True)
```

```python
for col in df.columns:
    print(col)
    display(df[col].unique())
```

```python
cat_cols = list(df.select_dtypes('O').columns)
cat_cols.append('household_smokers')
cat_cols.append('household_size')
```

```python
num_cols = list(df.columns)
for col in cat_cols:
    num_cols.remove(col)
```

```python
cat_cols.remove('depression')
print(len(df.columns))
print(len(cat_cols))
print(len(num_cols))
```

```python
print('Numerical Columns:\n')
print(num_cols)
print('\n\n Categorical Columns:\n')
print(cat_cols)
```

```python
fig = px.histogram(df, x="bronchitis_currently", color="depression", barmode="group", 
                   title="Depression By Gender")
fig.show()
```

```python
temp_df = df.groupby('bronchitis_currently')['depression'].value_counts(normalize=True)
temp_df = temp_df.mul(100).rename('percent').reset_index()
temp_df
```

```python
order_list = list(df['race'].unique())
order_list.sort()
order_list
```

```python
fig = px.histogram(temp_df, x="race", y='percent', color="depression", 
                   barmode="group", 
                   title="Percent Depression By Race")
fig.show()
```

```python
fig = px.histogram(temp_df[temp_df['depression']=='Depressed'], x="race", y='percent', 
                   barmode="group", 
                   title="Percent Depression By Race")
fig.show()
```
